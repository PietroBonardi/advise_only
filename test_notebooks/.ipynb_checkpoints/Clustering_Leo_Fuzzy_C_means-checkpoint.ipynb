# Introduzione Fuzzy C Means

L'obiettivo di questo esempio è quello di clusterizzare i dati forniti da _Advide Only_. 

In particolare, dobbiamo prendere i dati di un cliente, clusterizzare su questo, e capire come i vari clienti si dividano. In questo contesto, mi è venuta l'idea di andare a usare clustering **fuzzy**, ovvero dove non si associa un oggetto a un solo cluster, ma esso ha un **peso di appartenenza** a tutti i cluster proposti.
```python
def pre_processing():
#     Load data
    df = pd.read_csv("../dataset/DatasetClientClustering.csv", 
                 engine='c', 
                 sep=',', 
                 encoding='latin-1')
#     Drop the first 8 columns. They do not contain any data
    df = df.drop(df.columns[list(np.arange(8))], axis=1)

#     We do not use the ClientID for any analysis
    df = df.drop(columns=["ClientID"])

#     Division in the two main categories
    current_investment = ["PortfolioRisk", "PortfolioHorizon", "AuM", "BondInvestments", "EquityInvestments",
                         "MoneyMarketInvestments", "OtherInvestments", "Cash"]
    person_var = ["RiskPropension", "ClientInvestmentHorizon", "ClientKnowledgeExperience", "ClientPotentialIndex",
                 "IncomeHighLow", "Sex", "Age", "IncomeNeed", "LongTermCareNeed", "ProtectionNeed",
                 "InheritanceIndex", "PanicMood", "ClientDateStart", "NoTrustInBanks"]
#     creation of separate dataset
    df_person_var = df[person_var]
    df_current_investment = df[current_investment]
#     return as a toople of dataframes
    return(df_person_var, df_current_investment)
```

```python
df_person_var, _ = pre_processing()
```

# Primo tentativo di Clustering

Uso in questo contesto `RiskPropension` e `ClientPotentialIndex` come variabili per fare clustering. Si sono scelte queste due solamente per comodità. Infatti, inizialmente avevo pensato di utilizzare tecniche differenti, come l'uso di PCA per poter prendere la variazione di più variabili insieme, ma è qualcosa di delicato: sebbene da un lato questo permetta di considerare la variazione di più attributi insieme, possono esserci dei problemi per l'interpretabilità, in quanto non è sempre chiaro che cosa rappresentino le variabili. 

```python
def first_fuzzy_general(df = None):
    colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
    fig1, axes1 = plt.subplots(3, 3, figsize=(8, 8))
#     The data must be fed as a matrix
    alldata = np.vstack((df["RiskPropension"], 
                        df["ClientPotentialIndex"]))
    fpcs = []

    for ncenters, ax in enumerate(axes1.reshape(-1), 2):
        cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
            alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)

        # Store fpc values for later
        fpcs.append(fpc)

        # Plot assigned clusters, for each data point in training set
        cluster_membership = np.argmax(u, axis=0)
        for j in range(ncenters):
            ax.plot(df["RiskPropension"][cluster_membership == j],
                    df["ClientPotentialIndex"][cluster_membership == j], '.', color=colors[j])

        # Mark the center of each fuzzy cluster
        for pt in cntr:
            ax.plot(pt[0], pt[1], 'rs')

        ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc))
        ax.axis('off')

    fig1.tight_layout()
```

```python
cntr, u, u0, d, jm, p, fpc = first_fuzzy_general(df_person_var[["RiskPropension", "ClientPotentialIndex"]])
```

Da queste variabili non si evince nulla di particolare. Il numero di cluster che risulta più solido è `2`, ma anch'esso non sembra rappresentare nulla di concreto. In ogni caso, il fatto che non corrisponda a qualcosa di reale non è per forza negativo: il nostro obiettivo è identificare dai dati dei gruppi differenti. 

Decido quindi di prendere i centroidi del clustering con 2 elementi.
