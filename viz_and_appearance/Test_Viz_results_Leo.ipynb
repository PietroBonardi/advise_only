```python
# Load the dataset with labels
with open("../DatasetClientClustering_lables.csv", 'r') as file:
    labelled_all_data = pd.read_csv(file, engine='c')
    
# Load the variables used for the clustering
with open("../all_comparisons/config.yml", 'r') as file:
    load_vars = yaml.safe_load(file)
```

```python
X = StandardScaler().fit_transform(labelled_all_data[load_vars["clustering_variables"]])
print("-- TSNE manifold learning projection in 2D")
N_transform = 2
X_transformed = TSNE(n_components = N_transform).fit_transform(X)
```

## Variabili di clustering

### Proiezione 2D dei cluster

```python
cluster_palette_RGB = [tuple(
                            int(color.lstrip("#")[i:i+2], 16)/255
                             for i in (0, 2, 4)
                            ) 
                       for color in palette["clusters"]]

clusters_colors = [sns.desaturate(cluster_palette_RGB[col+1], sat)
                        if col >= 0 else (0.5, 0.5, 0.5)
                        for col, sat in zip(hdbscan_clusterer.labels_, hdbscan_clusterer.probabilities_)]

fig, ax = plt.subplots(figsize=(10,10))
scatter = ax.scatter(x = X_transformed[:,0], y = X_transformed[:,1], c = clusters_colors)
lines = [Line2D([0], [0], color=c, linewidth=3, linestyle='', marker='o')
         for i, c in enumerate(palette['clusters'])]
labels = ["noise", 0,1,2,3]
plt.legend(lines, labels, title="Clusters")
plt.title("TSNE Manifold projection for the clustering")
plt.xlabel("X")
plt.ylabel("X")
plt.savefig("plots/TSNE_Projection.png")
plt.show()
```

```python
other_vars = ['Age', 'IncomeNeed', 'InheritanceIndex']
for other_var in other_vars:
    aux_df = np.asarray(labelled_all_data[['ProtectionNeed', other_var ]])
    cluster_palette_RGB = [tuple(
                                int(color.lstrip("#")[i:i+2], 16)/255
                                 for i in (0, 2, 4)
                                ) 
                           for color in palette["clusters"]]

    clusters_colors = [sns.desaturate(cluster_palette_RGB[col+1], sat)
                            if col >= 0 else (0.5, 0.5, 0.5)
                            for col, sat in zip(hdbscan_clusterer.labels_, hdbscan_clusterer.probabilities_)]

    fig, ax = plt.subplots(figsize=(10,10))
    scatter = ax.scatter(x = aux_df[:,0], y = aux_df[:,1], c = clusters_colors)
    lines = [Line2D([0], [0], color=c, linewidth=3, linestyle='', marker='o') if i < 4 
             else  Line2D([0], [0], color="#7F7F7F", linewidth=3, linestyle='', marker='o')
             for i, c in enumerate(palette['clusters'])]
    labels = ["noise",0,1,2,3]
    plt.legend(lines, labels, title="Clusters")
    plt.title("TSNE Manifold projection for the clustering")
    plt.xlabel('ProtectionNeed')
    plt.ylabel(other_var)
    plt.savefig("plots/scatter_ProtectioNeed_"+str(other_var)+".png")
    plt.show()
```

Da questi boxplot si vede come tutte le variabili utilizzate posseggano un qualche grado di clusterizzazione, come ci si potrebbe aspettare data la selezione delle stesse tramite entropia. In particolare, `ProtectionNeed` separa molto i cluster 0 e 1 dai cluster 2 e 3, tanto che, a meno di qualche valore outlier nel cluster 0, se un individuo ha un alto valore di `ProtectionNeed` sar sicuramente nei cluster 2 o 3. 
Per le altre variabili, sebbene sia presente una clusterizzazione, essa  meno netta che in questo caso. Per esempio, un individuo con alto valore di `IncomeNeed` (sopra 0.6) sar sicuramente nel cluster 1, mentre per bassi valori si trover in uno degli altri cluster. Ancora, nessun individuo sopra i 60 (circa) pu essere nel cluster 2 o 3, mentre il cluster 0 contiene un range molto ampio di et. 

Analoghi ragionamenti valgono per le due variabili non citate, ovvero `InheritanceIndex` e `RiskPropension`. 

Questi risultati mostrano lungo quali attributi avvenga la pi forte clusterizzazione, e quale sia il range dei diversi cluster lungo le varie dimensioni. Ovviamente, gli esempi e i ragionamenti fatti escludono il cluster -1, che rappresenta i punti di *noise*; su questi, si pu affermare che non ci sia una vera clusterizzazione, in quanto presentano valori completamente eterogenei per tutti gli attributi.

```python
clustering_variables_2 = load_vars["clustering_variables"].copy()
clustering_variables_2.append("Clusters")
```

```python
labelled_all_data[clustering_variables_2].groupby("Clusters").mean()
```

- **Cluster 0**: presenta un'*et* media di 58 e un ampio range; presenta una *income need* medio bassa, con quasi nessun valore sopra 0.5; ha un ampio range di *risk propension*, con una media poco sotto lo 0.5; i valori di *protection need* sono bassi, con quasi nessun individui avente sopra lo 0.5; presenta un ampio range di *inheritance index*, con media quasi esattamente a met. Da questi dati, si pu dire che questo cluster  caratterizzato da uomini intorno ai 60 anni, con poco bisogno di cedole/investimenti per il loro reddito (ovvero che possiedono altre fonti di reddito fisse), tendono a non essere troppo propensi al rischio, non hanno troppo bisogno di proteggere il proprio capitale dalle perdite del mercato e hanno un certo bisogni di ottimizzazione del loro partimonio per figli/eredi. 

    A oggi, questo gruppo tende in media ad avere un portofolio abbastanza rischioso, con molti dei soldi investiti in Bond e Equity; essi non tengono invece per niente soldi in Cash.
    
    Visto questo, si potrebbe suggerire di modificare leggermente gli investimenti all'interno di questo gruppo; sebbene da un lato siano in grado di supportare delle perdite, la tendenza a questi individui a non andare verso il rischio potrebbe far pensare di dare loro investimenti pi sicuri delle azioni: aumentare gli investimenti in Bond, Cash e, eventualmente il cliente avesse delle scadenze brevi oppure non volesse avere troppa liquidit, si potrebbe investire su Money Market.
    
    *considerazioni insieme*: investire maggiormente in bond, in quanto pi a lungo termine e pi in linea con le caratteristiche degli individui. Diminuire l'investimento in azioni. 
    
- **Cluster 1**:  il cluster con pi persone anziane, di et media 66, sebbene sia presenti persone dai 40 ai 100 anni. A differenza del cluster precedente, questi individui sono caratterizzati da una pi grande necessit di reddito da parte di cedole/dividendi; la loro propensione al rischio  la pi bassa tra tutti i cluster e, come il precedente, hanno una bassa *protection need*; come ci si potrebbe aspettare, il valore di *inheritance index*  il pi alto tra tutti i cluster.

    A oggi, questo gruppo  quello che presenta minore percentuale di liquidit, con la maggior parte dei soldi investiti in Bond. 
    
    Data la bassa propensione al rischio e la necessit di ottimizzare il patrimonio, l'attuale posizione di investimento potrebbe essere modificata aumentando la liquidit. Sebbene questo sia l'investimento meno redditizio,  anche il pi sicuro, in linea con le tendenze di questo gruppo; inoltre, mantenere i soldi in cash  sicuramente in ottomo ottimizzatore del patrimonio per gli eredi.
    
    *considerazioni insieme*: questo gruppo va bene, magari diminuire le azioni e aumentare cash, in linea con le caratteristiche psicologiche.
    
- **Cluster 2**: questo  il gruppo pi giovane, con et media di 27, sebbene siano presenti anche individui pi vecchi o pi giovani. Questi individui sono i pi propensi al rischio, hanno meno necessit di reddito da dividendi/cedole ma, probabilmente a causa di una minore disponibilit finanziaria, hanno bisogno di pi protezione finanziara da pardite. Inoltre, i valori di ottimizzazione per gli eredi sono molto bassi.

    Il loro investimento attuale potrebbe essere modificato puntando a investire una maggior percentuale del loro asset in Bond da Cash. Sebbene questo gruppo sia meno in grado di sopportare delle perdite, una parte maggiore del denaro potrebbe essere investita in Azioni o Moneta, in linea con le tendeze di rischio.
    
    *considerazioni insieme*: consigliamo investimenti pi a lungo termine, nelle possibilit del cliente. Togliere cash e investire in qualcosa, e.g. bond, azioni o moneta. Se impossibililati da investire in bond per limitazioni in cash, investire moneta.
    
- **Cluster 3**: il gruppo dei *middle age*, il range delle et vai tra i 30 e i 60, con una media di 45. In linea con il cluster precedente, presentano poco *IncomeNeed* e una *RiskPropension* abbastanza elevata. La *ProtectionNeed*  statisticamente uguale al cluster 2 ma, a differenza di questo, il valore di *InheritanceIndex*  quasi 0.5, ovvero pi simile al cluster 0.

    L'investimento di questo cluster  sicuramente il pi *au pair* con le loro caratteristiche, sebbene non presentino un elevato rischio. Una bassa capacit di supportare perdite finanziare, ma una tendenza al rischio media, farebbe infatte puntare su investimenti come i Bond, fatto che si vede essere vero. L'unico consiglio che si potrebbe dare  quello di investire pi soldi da Cash in Azioni, in linea con le tendenze al rischio.
